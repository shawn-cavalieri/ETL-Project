```python
url = 'https://www.fangraphs.com/leaders.aspx?pos=all&stats=bat&lg=all&qual=0&type=8&season=2019&month=0&season1=2019&ind=0&team=0,ts&rost=&age=&filter=&players=0'

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html.parser')
```

```python
soup.find_all()
```

```python
soup.find_all('tr', attrs = {'class': 'rgAltRow'})
```

```python
row = soup.find('tr', attrs = {'class': 'rgAltRow'})
for data in row.find_all('td'):
    print(data.get_text())
```

```python
header = soup.find_all('th', attrs ={'class': 'rgHeader'})

columns = []
for row in header:
    print(row.text)
    columns.append(row.text)

columns

final_df = pd.DataFrame(columns=columns)

final_df
```
